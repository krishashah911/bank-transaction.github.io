The project task is as follows:

### Stage 1 : User Validity

- Ask user to enter Full Name or User ID
- If Full Name / User ID not there then tell invalid User
- If it exists, then ask user to enter Bank Account Number
- If A/c number doesn't exist, then tell invalid user
- If both are correct, then proceed to Stage 2
- Give user three tries. If invalid input is given in all three tries then change account status from Active to Blocked and print Account Blocked and print quit program in Stage 1

### Stage 2: OTP Generation and Validity

- Through code generate a 4 digit random OTP number
- Send this OTP to that User via email
- Then ask user to enter OTP from mail. Match this OTP with one generated in code.
- If OTP matches then go to Stage 3. Else say Invalid OTP. 
- Give user three tries. If wrong OTP in all three tries then change account status from Active to Blocked and print Account Blocked and print quit program in Stage 2

### Stage 3 : Choice of Transaction

- Ask user which type of transaction he wishes to do. ‘C’ for credit money and ‘D’ for debit money
- If User enters any other input here then handle error and print Invalid Input. Again give user three tries and if all wrong then change account status from Active to Blocked and print Account Blocked and print quit program in Stage 3
- If correct input then print current balance
- Then ask user for amount to be credited or debited
- Add or subtract amount entered by user to the current account balance and print the final account balance and write the same to account balance cells in excel workbook.
- Print “Successful Transaction of Rs.______ completed”
- **BONUS QUESTION: Print a proper mini-statement of the balance and display it. (2 points)**

```python
import pandas as pd
import openpyxl
from openpyxl import load_workbook
wb = load_workbook('SITIP19-VIT-Python-FinalProject1-Workbook.xlsx')

# print(wb.sheetnames)
sh1=wb['UserDetails']

Sr_list=[]
Id_list=[]
Ac_list=[]
Name_list=[]
Email_list=[]
Balance_list=[]
Status_list=[]

df1 = pd.DataFrame(columns = [sh1['A1'].value,sh1['B1'].value,sh1['C1'].value,sh1['D1'].value,sh1['E1'].value,sh1['F1'].value,sh1['G1'].value])

for i in range(2,17):
    df1.loc[i-2] = [sh1.cell(row=i,column=1).value , sh1.cell(row=i,column=2).value , sh1.cell(row=i,column=3).value , sh1.cell(row=i,column=4).value , sh1.cell(row=i,column=5).value , sh1.cell(row=i,column=6).value , sh1.cell(row=i,column=7).value]
    
    Sr_list.append(sh1.cell(row=i,column=1).value)
    Id_list.append(sh1.cell(row=i,column=2).value)
    Ac_list.append(sh1.cell(row=i,column=3).value)
    Name_list.append(sh1.cell(row=i,column=4).value)
    Email_list.append(sh1.cell(row=i,column=5).value)
    Balance_list.append(sh1.cell(row=i,column=6).value)
    Status_list.append(sh1.cell(row=i,column=7).value)
```

```python
from datetime import datetime
def bonus(i):

    acac = str(Ac_list[i])
    print("")
    print("")
    print("*********************************************")
    
    print("NIRJULI 2ND ATM".center(50))
    print("")
    
    print("DATE".ljust(15),end="")
    print("TIME".center(15),end="")
    print("TERMINAL".rjust(15))
    print(datetime.now().strftime("%x").ljust(15),end="")
    print(datetime.now().strftime("%H:%M:%S").center(15),end="")
    print("S10A009535002".rjust(15))
    print("")
    print("NAME       : ",Name_list[i])
    print("ID NUMBER  : ",Id_list[i])
    print("A/C NUMBER : ",acac[:3]+"XXXXXXXXXXX"+acac[-3:])
    print("")
    print("AVAIL BAL  : ",Balance_list[i])
    
    print("*********************************************")
```

## Stage 3

```python
def stage3(i):
    cc=0
    print("##### 'C' for Credit Money or 'D' for Debit Money #####")
    while cc<3:
        choice=input("Enter your choice: ")
    
        if choice == 'C':
            print("Current Balance: ",Balance_list[i])
            credit = int(input("Enter amount to be credited- "))
            final=Balance_list[i]+credit
            print("\nTotal Balance = ",final)
            Balance_list[i]=final
            sh1['F'+str(i+2)] = final
            wb.save('SITIP19-VIT-Python-FinalProject1-Workbook.xlsx') 
            print("Successful transaction of Rs.",credit,"completed")
            cc=4
            bonus(i)
        elif choice == 'D':
            print("Current Balance: ",Balance_list[i])
            debit = int(input("Enter amount to be debited- "))
            final=Balance_list[i]-debit
            if final>=0:
                print("\nTotal Balance = ",final)
                Balance_list[i]=final
                sh1['F'+str(i+2)] = final
                wb.save('SITIP19-VIT-Python-FinalProject1-Workbook.xlsx') 
                print("Successful transaction of Rs.",debit,"completed")
                bonus(i)
            else:
                print("Account does not have a balance of Rs.",debit)
            cc=4
        else:
            print("*Invalid Input*")
            cc+=1
            
    if cc==3:
        Status_list[i]='Blocked'
        sh1['G'+str(i+2)] = "Blocked"
        wb.save('SITIP19-VIT-Python-FinalProject1-Workbook.xlsx') 
        print("\nYour Acount is blocked")
        print("Quit Program in Stage 3\n")
   
```

## Stage 2

```python
import random 
def stage2(i):
    ss=0
    while ss<3:
        
        otp = random.randint(1000,9999)
    
        import smtplib
        from email.mime.multipart import MIMEMultipart
        from email.mime.text import MIMEText

        fromadd = "your email Id"
        toadd = Email_list[i]
        msg = MIMEMultipart()
        msg['From'] = fromadd
        msg['To'] = toadd
        msg['Subject'] = "One Time Password (OTP)"
        body = "Your OTP for account is "+str(otp)
        msg.attach(MIMEText(body,'plain'))
        server = smtplib.SMTP('smtp.gmail.com', port=587)
        server.starttls()
        server.login(fromadd,"your password")
        text = msg.as_string()
        server.sendmail(fromadd,toadd,text)
        server.quit()
    
        OTP = int(input("Enter received OTP: "))
    
        if OTP == otp:
            ss=4
            stage3(i)
        else:
            print("*Invalid OTP*")
            ss+=1
    
    if ss==3:
        Status_list[i]='Blocked'
        sh1['G'+str(i+2)] = "Blocked"
        wb.save('SITIP19-VIT-Python-FinalProject1-Workbook.xlsx') 
        print("\nYour Acount is blocked")
        print("Quit Program in Stage 2\n")
```

## Stage 1

```python
def stage1():
    rr=0
    kk=0
    while kk<3 and rr<3:

        x = int(input("Enter User ID: "))
        if x in Id_list:
            input_ac = input("Enter Account Number: ")
            if input_ac in Ac_list:
                for i in range(len(Id_list)):
                    if x == Id_list[i] and input_ac == Ac_list[i]:
                        kk=3
                        stage2(i)
                        
            else:
                print("*Invalid Account*")
                rr+=1
        else:
            print("*Invalid ID*")
            kk+=1
        
# print("kk",kk)
# print("rr",rr)
    if rr==3:
        for j in range(len(Id_list)):
            if x == Id_list[j]:
                Status_list[j]='Blocked'
                sh1['G'+str(j+2)] = "Blocked"
                wb.save('SITIP19-VIT-Python-FinalProject1-Workbook.xlsx') 
        print("\nYour Acount is blocked")
        print("Quit Program in Stage 1\n")

stage1()
```
